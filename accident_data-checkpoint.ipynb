```python
#fix null values
#1-fix null values in numerical data
numerical_columns=['Age_of_Vehicle','Driver_IMD_Decile','Engine_Capacity_.CC.','Vehicle_Location.Restricted_Lane']
for col in numerical_columns:
    vehicle_data[col]=vehicle_data[col].median()
#2-fix null values in categorical data
categorical_columns=['Hit_Object_in_Carriageway','Hit_Object_off_Carriageway','make','model','Propulsion_Code','Skidding_and_Overturning']
for col in categorical_columns:
    vehicle_data[col]=vehicle_data[col].mode()[0]
```

```python
#merge two tables
merged_df=pd.merge(accident_data,vehicle_data,on='Accident_Index',how='left')
```

Question 1 : What are the locations with the highest number of casualties and vehicles involved?

```python
location_summary=merged_df.groupby('Local_Authority_(District)')[['Number_of_Casualties','Number_of_Vehicles']].sum().sort_values(by='Number_of_Casualties',ascending=False).head(10)
plt.figure(figsize=(12,6))
location_summary.plot(kind='bar',figsize=(12,6),width=0.7,colormap='Pastel1')
plt.title=('Top 10 Locations by Number of Casualties and Vehicles')
plt.xlabel('Local_Authority_(District)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

Question 2 : For each vehicle type, How many number of casualties?

```python
vehicle_casualties=merged_df.groupby('Vehicle_Type')['Number_of_Casualties'].sum().sort_values(ascending=False)
top3=vehicle_casualties.head(3)
bottom3=vehicle_casualties.tail(3)
fig, axes=plt.subplots(1,2,figsize=(12,6))
axes[0].pie(top3.values,labels=top3.index,autopct='%1.1f%%',startangle=90,colors=sns.set_palette('Set2',len(vehicle_casualties)))
axes[0].set_title('Top 3 Vehicle Types by Casualties')
axes[1].pie(bottom3.values,labels=bottom3.index,autopct='%1.1f%%',startangle=90,colors=sns.set_palette('Set3',len(vehicle_casualties)))
axes[1].set_title('Bottom 3 Vehicle Types by Casualties')
plt.tight_layout()
plt.show()
```

Severity across top 7 local authorities

```python
la_severity =merged_df.pivot_table(index='Local_Authority_(District)', columns='Accident_Severity', aggfunc='size', fill_value=0)
top_local_auth = la_severity.sum(axis=1).sort_values(ascending=False).head(7).index
Top_la_severity = la_severity.loc[top_local_auth]
Top_la_severity.plot(kind='bar', figsize=(10, 6), color=sns.color_palette('Pastel2'))
plt.title=("Accident Severity for Top 7 Local Authorities")
plt.xlabel("Local Authority")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.legend(title="Severity")
plt.tight_layout()
plt.show()
```

Casualties by Driver Age Band and Sex

```python
pivot = merged_df.pivot_table(values='Number_of_Casualties',
                       index='Age_Band_of_Driver',
                       columns='Sex_of_Driver',
                       aggfunc='sum')
pivot = pivot.sort_index()
sns.color_palette('Dark2')
pivot.plot(kind='line', marker='o', figsize=(12, 7))
plt.title=('Casualties by Driver Age Band and Sex')
plt.xlabel('Driver Age Band')
plt.ylabel('Number of Casualties')
plt.legend(title='Sex of Driver')
plt.tight_layout()
plt.show()
```

Top 10 Locations by Casualties and Vehicles 

```python
Top_locations = merged_df.groupby('Local_Authority_(District)').agg({
    'Number_of_Casualties': 'sum',
    'Number_of_Vehicles': 'sum'
}).sort_values(by='Number_of_Casualties', ascending=False).head(10)
sns.color_palette('Set2')
Top_locations.plot(kind='barh', stacked=True, figsize=(12, 7))
plt.title=('Top 10 Districts by Casualties and Vehicles Involved')
plt.xlabel('Total Count')
plt.ylabel('Local Authority (District)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

Casualties by Day of the Week 

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
Casualties_by_day = merged_df.groupby('Day_of_Week')['Number_of_Casualties'].sum().reindex(day_order)
Casualties_by_day.plot(kind='area', figsize=(12, 6), alpha=0.7)
sns.color_palette('Pastel2')
plt.title=('Total Casualties by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Casualties')
plt.tight_layout()
plt.show()
```

Group by Speed Limit and Accident Severity

```python
severity_speed = merged_df.groupby(['Speed_limit', 'Accident_Severity']).size().unstack(fill_value=0)
plt.figure(figsize=(10, 6))
sns.heatmap(severity_speed, annot=True, fmt='d', cmap='crest', linewidths=0.5)
plt.title=('Heatmap of Accident Severity by Speed Limit')
plt.xlabel('Accident Severity')
plt.ylabel('Speed Limit')
plt.tight_layout()
plt.show()
```
